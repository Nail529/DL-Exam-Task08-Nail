```python
import torch
import torch.nn as nn
import torch.optim as optim
from torchvision import datasets, models, transforms
from torch.utils.data import DataLoader, Subset
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
from sklearn.metrics import confusion_matrix
import time
import pandas as pd

#Seed
SEED = 20240108
torch.manual_seed(SEED)
np.random.seed(SEED)
if torch.cuda.is_available():
    torch.cuda.manual_seed(SEED)

#DATA PREPROCESSING
transform = transforms.Compose([
    transforms.Resize((224, 224)),
    transforms.ToTensor(),
    transforms.Normalize((0.485, 0.456, 0.406), (0.229, 0.224, 0.225))
])
```

```python
#Here i download Food101 dataset
full_train_data = datasets.Food101(root='data', split='train', download=True, transform=transform)
full_test_data = datasets.Food101(root='data', split='test', download=True, transform=transform)

# I select 5 classes
chosen_classes = ['pizza', 'hamburger', 'sushi', 'tacos', 'ice_cream']
class_to_idx = {full_train_data.classes[i]: i for i in range(len(full_train_data.classes)) if full_train_data.classes[i] in chosen_classes}
target_indices = [class_to_idx[c] for c in chosen_classes]

# Now i filter our datasets
def filter_indices(dataset, target_indices):
    return [i for i, label in enumerate(dataset._labels) if label in target_indices]

train_subset = Subset(full_train_data, filter_indices(full_train_data, target_indices))
test_subset = Subset(full_test_data, filter_indices(full_test_data, target_indices))

# And Create Loaders
train_loader = DataLoader(train_subset, batch_size=32, shuffle=True)
test_loader = DataLoader(test_subset, batch_size=32, shuffle=False)

# Now i am mapping labels to 0-4 for my 5 classifer
label_map = {old_idx: new_idx for new_idx, old_idx in enumerate(target_indices)}
```

```python
def get_model():
    model = models.resnet18(weights='IMAGENET1K_V1')

    # Stopping Layers
    for param in model.parameters():
        param.requires_grad = False

    # then replacing final layer
    model.fc = nn.Linear(512, 5) # 5 classes
    return model
```

```python
def train(model, loader, criterion, optimizer, device):
    model.train()
    running_loss = 0.0
    correct = 0
    total = 0

    for images, labels in loader:
        # Mapping Labels
        labels = torch.tensor([label_map[l.item()] for l in labels]).to(device)
        images = images.to(device)

        optimizer.zero_grad()
        outputs = model(images)
        loss = criterion(outputs, labels)
        loss.backward()
        optimizer.step()

        running_loss += loss.item()
        _, predicted = outputs.max(1)
        total += labels.size(0)
        correct += predicted.eq(labels).sum().item()

    return running_loss/len(loader), 100.*correct/total

def evaluate(model, loader, device):
    model.eval()
    correct = 0
    total = 0
    with torch.no_grad():
        for images, labels in loader:
            labels = torch.tensor([label_map[l.item()] for l in labels]).to(device)
            images = images.to(device)
            outputs = model(images)
            _, predicted = outputs.max(1)
            total += labels.size(0)
            correct += predicted.eq(labels).sum().item()
    return 100.*correct/total
```

```python
# 1. Setup Label Map
chosen_classes = ['pizza', 'hamburger', 'sushi', 'tacos', 'ice_cream']
# Mapping Original Class jndices
label_map = {full_train_data.class_to_idx[cls]: i for i, cls in enumerate(chosen_classes)}

device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
criterion = nn.CrossEntropyLoss()

# VERSION 1: Adam
print("--- Starting Version 1: Adam Optimizer ---")
#ResNet18 Function
model_v1 = get_model().to(device)
optimizer_v1 = optim.Adam(model_v1.fc.parameters(), lr=0.0001)

v1_acc_history = []
v1_loss_history = []

for epoch in range(10):
    loss, acc = train(model_v1, train_loader, criterion, optimizer_v1, device)
    val_acc = evaluate(model_v1, test_loader, device)

    v1_loss_history.append(loss)
    v1_acc_history.append(val_acc)
    print(f"Epoch {epoch+1}/10 - Loss: {loss:.4f}, Train Acc: {acc:.2f}%, Val Acc: {val_acc:.2f}%")
```

```python
# VERSION 2: SGD
print("\n--- Starting Version 2: SGD with Momentum ---")
model_v2 = get_model().to(device)
optimizer_v2 = optim.SGD(model_v2.fc.parameters(), lr=0.001, momentum=0.9)

v2_acc_history = []
v2_loss_history = []

for epoch in range(10):
    loss, acc = train(model_v2, train_loader, criterion, optimizer_v2, device)
    val_acc = evaluate(model_v2, test_loader, device)

    v2_loss_history.append(loss)
    v2_acc_history.append(val_acc)
    print(f"Epoch {epoch+1}/10 - Loss: {loss:.4f}, Train Acc: {acc:.2f}%, Val Acc: {val_acc:.2f}%")
```

```python
#  Comparing Plot
plt.figure(figsize=(10, 5))
plt.plot(range(1, 11), v1_acc_history, label='Version 1: Adam', marker='o')
plt.plot(range(1, 11), v2_acc_history, label='Version 2: SGD', marker='s')
plt.title('Validation Accuracy Comparison: Adam vs SGD')
plt.xlabel('Epochs')
plt.ylabel('Accuracy (%)')
plt.legend()
plt.grid(True)
plt.savefig('comparison_plot.png')
plt.show()

#  Comparing Table
data = {
    "Metric": ["Best Val Accuracy", "Final Training Loss"],
    "Version 1 (Adam)": [f"{max(v1_acc_history):.2f}%", f"{v1_loss_history[-1]:.4f}"],
    "Version 2 (SGD)": [f"{max(v2_acc_history):.2f}%", f"{v2_loss_history[-1]:.4f}"]
}
print("\n--- Comparison Table ---")
print(pd.DataFrame(data))
```

```python
# Selecting best model
best_model = model_v1 if max(v1_acc_history) > max(v2_acc_history) else model_v2

# Confusion Matrix
all_preds = []
all_true = []
best_model.eval()

with torch.no_grad():
    for images, labels in test_loader:
        mapped_labels = torch.tensor([label_map[l.item()] for l in labels]).to(device)
        outputs = best_model(images.to(device))
        _, preds = outputs.max(1)
        all_preds.extend(preds.cpu().numpy())
        all_true.extend(mapped_labels.cpu().numpy())

cm = confusion_matrix(all_true, all_preds)
plt.figure(figsize=(8, 6))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', xticklabels=chosen_classes, yticklabels=chosen_classes)
plt.title('Confusion Matrix - Best Model')
plt.ylabel('Actual Label')
plt.xlabel('Predicted Label')
plt.savefig('confusion_matrix.png')
plt.show()

#  10  Predictions for Example
plt.figure(figsize=(15, 6))
images, labels = next(iter(test_loader))
outputs = best_model(images.to(device))
_, preds = outputs.max(1)
mapped_labels = [label_map[l.item()] for l in labels]

for i in range(10):
    plt.subplot(2, 5, i+1)
    img = images[i].permute(1, 2, 0).numpy()
    img = np.clip(img * [0.229, 0.224, 0.225] + [0.485, 0.456, 0.406], 0, 1)
    plt.imshow(img)
    color = 'green' if preds[i] == mapped_labels[i] else 'red'
    plt.title(f"True: {chosen_classes[mapped_labels[i]]}\nPred: {chosen_classes[preds[i]]}", color=color)
    plt.axis('off')
plt.savefig('sample_predictions.png')
plt.show()
```
